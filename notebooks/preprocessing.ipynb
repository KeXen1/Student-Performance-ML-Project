```python
# preprocessing.ipynb
# Load raw UCI Student Performance dataset, clean it, and save the processed version
import pandas as pd
import numpy as np
import os
```

```python
# Clone repo
!git clone https://github.com/KeXen1/Student-Performance-ML-Project.git

%cd Student-Performance-ML-Project
```

```python
# Load raw dataset
# UCI file uses semicolon as the separator
raw = pd.read_csv("data/raw/student-mat.csv", sep=";")

display(raw.head())
print("Raw shape:", raw.shape)
```

```python
# Inspect data types and column names
print(raw.dtypes)
print()
print("Columns:")
print(raw.columns.tolist())
```

```python
# Check for missing values
missing = raw.isnull().sum()
print("Total missing values:", missing.sum())
print()
print(missing[missing > 0])
```

```python
# Handle missing values
# Numeric columns: fill with median
# Categorical columns: fill with mode
data = raw.copy()

numeric_cols = data.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = data.select_dtypes(include=["object"]).columns

for col in numeric_cols:
    if data[col].isnull().any():
        data[col] = data[col].fillna(data[col].median())

for col in categorical_cols:
    if data[col].isnull().any():
        data[col] = data[col].fillna(data[col].mode()[0])

print("Missing values after handling:", data.isnull().sum().sum())
```

```python
# Drop duplicate rows
before = data.shape[0]
data = data.drop_duplicates().reset_index(drop=True)
after = data.shape[0]

print("Rows before:", before)
print("Rows after:", after)
print("Duplicates removed:", before - after)
```

```python
# Verify grade ranges (should be 0-20)
for col in ["G1", "G2", "G3"]:
    print(col, "range:", data[col].min(), "to", data[col].max())
```

```python
# Summary statistics
display(data.describe())
```

```python
# Distribution of categorical features
for col in categorical_cols:
    print(col)
    print(data[col].value_counts())
    print()
```

```python
# Save cleaned dataset
# Categorical variables are kept as strings
# Encoding and scaling are handled later inside the model pipeline
os.makedirs("data/processed", exist_ok=True)

data.to_csv("data/processed/cleaned_data.csv", index=False)

print("Cleaned dataset saved to data/processed/cleaned_data.csv")
print("Final shape:", data.shape)
```
